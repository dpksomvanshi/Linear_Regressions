### Establish Relatioship in Profit vs RND

## Profit ~ RND
## profit = B0+B1*RND 

```python
# visualizing RND and Profit
import matplotlib.pyplot as plt
plt.scatter(df['RND'],df['PROFIT'])
plt.xlabel('RND')
plt.ylabel('PROFIT')
plt.title('RND vs PROFIT')
plt.grid()
plt.show()
```

## seperate dependant and independant features

```python
X=df[['RND']]
Y=df[['PROFIT']]
```

```python
X
```

```python
Y
```

```python
X.head()
```

```python
Y.head()
```

## Build a Linear Regression model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
```

```python
model.intercept_
```

```python
model.coef_
```

## Equation for PROFIT Vs RND
~~~
PROFIT = 49032.89 + 0.8542*RND
~~~

```python
ypred = model.predict(X)
```

```python
ypred
```

### Visualising RND and PROFIT and line

```python
import matplotlib.pyplot as plt
plt.scatter(df['RND'], df['PROFIT'])
plt.plot(X,ypred,c='red')
plt.xlabel('RND')
plt.ylabel('PROFIT')
plt.title('RND vs PROFIT')
plt.grid()
plt.show()
```

```python
import seaborn as sns
sns.regplot(data=df,x='RND',y='PROFIT')
```

# Evaluate model
1. Mean Squared Error
2. Root Mean Squared Error
3. Mean Absolute Error
4. R2 Score

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
MSE=mean_squared_error(Y,ypred)
RMSE = MSE**(1/2)
print(f'Mean Squared Error: {MSE:.2f}')
print(f'Root Mean Squared Error: {RMSE: .2f}')
```

```python
MAE = mean_absolute_error(Y,ypred)
print(f'Mean Absolute Error: {MAE:.2f}')
```

```python
R2_Score = r2_score(Y,ypred)
print(f'R2 Score: {R2_Score: .4f}')
```
